# Algumas Funções Básicas de Lista

## Tamanho da Lista

tamanho = len(lista)

```python
produtos = ['apple tv', 'mac', 'iphone x', 'IPad', 'apple watch', 'mac book', 'airpods']
```

- Quantos produtos temos a venda?

```python
tamanho_lista = len(produtos)
print(f'Temos um total de {tamanho_lista} produtos cadastrado na lista')
```

## Maior e Menor Valor

maior = max(lista)

menor = min(lista)

```python
vendas = [1000, 1500, 15000, 270, 900, 100, 1200]
```

- Qual o item mais vendido?
- Qual o item menos vendido?

```python
mais_vendido = max(vendas)
menos_vendido = min(vendas)

print(f'O produto {produtos[2]} mais vendido teve {mais_vendido} unidades vendidas e o {produtos[5]} menos vendido teve {menos_vendido} unidades vendidas')
```

```python
i = vendas.index(mais_vendido)
produto_mais_vendido = produtos[i]
print(f'O produto mais vendido foi {produto_mais_vendido} e a quantidade de {vendas[i]} unidades')

i = vendas.index(menos_vendido)
produto_menos_vendido = produtos[i]
print(f'O produto menos vendido foi {produto_menos_vendido} e a quantidade de {vendas[i]} unidades')
```
