#### 8. Vamos criar um conversor de temperatura. Faça um Programa que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius.
$C = \frac{5}{9}(F-32)$

```python
temperatura_farenheit = float(input("Insira a temperatura em Fahrenheit: "))
temperatura_celsius = (temperatura_farenheit - 32) * 5 / 9
print(f"A temperatura em Celsius é: {temperatura_celsius:.2f} °C")
```

#### 9. Faça um Programa que peça a temperatura em graus Celsius, transforme e mostre em graus Fahrenheit.
$F = \frac{9}{5}C + 32$

```python
temperatura_celsius = float(input("Insira a temperatura em Celsius: "))
temperatura_farenheit = (temperatura_celsius * 9 / 5) + 32
print(f"A temperatura em Fahrenheit é: {temperatura_farenheit:.2f} °F")
```

#### 11. Tendo como dado de entrada a altura (h) de uma pessoa, construa um algoritmo que calcule seu peso ideal, utilizando as seguintes fórmulas:
##### a. Para homens: $P = 72,7h - 58$
##### b. Para mulheres: $P = 62,1h - 44,7$

```python
altura_homens = float(input("Insira a altura do homem em metros: "))
altura_mulheres = float(input("Insira a altura da mulher em metros: "))
peso_ideal_homens = (72.7 * altura_homens) - 58
peso_ideal_mulheres = (62.1 * altura_mulheres) - 44.7
print(f"O peso ideal para a altura de {altura_homens} metros é: {peso_ideal_homens:.2f} kg")
print(f"O peso ideal para a altura de {altura_mulheres} metros é: {peso_ideal_mulheres:.2f} kg")
```

#### 12. Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.

```python
ganho_hora = float(input("Insira o quanto você ganha por hora: R$ "))
horas_trabalhadas = int(input("Insira o número de horas trabalhadas no mês: "))
```

#####  Calcule o salário bruto (horas * salario por hora)

```python
salario_bruto = ganho_hora * horas_trabalhadas
```

##### Calcule o desconto do IR (11% do salário bruto)

```python
imposto_de_renda = salario_bruto * 0.11
```

##### Calcule o desconto do INSS (8% do salário bruto)

```python
inss = salario_bruto * 0.08
```

##### Calcule o desconto do sindicato (5% do salário bruto)

```python
desconto_sindical = salario_bruto * 0.05
```

##### Calcule o salário líquido (salário bruto - descontos)

```python
salario_liquido = salario_bruto - (imposto_de_renda + inss + desconto_sindical)
print(f"O salário bruto é: R$ {salario_bruto:.2f}")
print(f"O salário líquido é: R$ {salario_liquido:.2f}")
```

#### 13. Faça um programa para uma loja de tintas. O programa deverá pedir o tamanho em metros quadrados da área a ser pintada. Considere que a cobertura da tinta é de 1 litro para cada 3 metros quadrados e que a tinta é vendida em latas de 18 litros, que custam R\$ 80,00. Informe ao usuário a quantidades de latas de tinta a serem compradas e o preço total. (para simplificação nesse momento, não se preocupe em arredondar a quantidade de latas a serem compradas - vamos trabalhar isso em breve)

```python
tamanho_area_pintada = float(input("Insira o tamanho da área a ser pintada em metros quadrados: "))

cobertura_litro = tamanho_area_pintada / 3 # m² por litro
litros_por_lata = cobertura_litro / 18  # 18 litros por lata
preco = litros_por_lata * 80

print(f"O número de latas necessárias é: {litros_por_lata:.2f}")
print(f"O custo total da tinta é: R$ {preco:.2f}")
```

#### 14. Faça um programa que peça o tamanho de um arquivo para download (em MB) e a velocidade de um link de Internet (em Mbps), calcule e informe o tempo aproximado de download do arquivo usando este link (em minutos).

Detalhe: MB significa megabyte, Mb (com b minúsculo) significa megabit. Um megabit é 1/8 de um megabyte. 

```python
arquivo = float(input("Insira o tamanho do arquivo em MB: "))
velocidade_internet = float(input("Insira a velocidade da sua internet em Mbps: "))
tempo_segundos = (arquivo * 8) / velocidade_internet
tempo_minutos = tempo_segundos / 60
print(f"O tempo aproximado para download é: {tempo_minutos:.2f} minutos")
```
