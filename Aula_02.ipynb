# Misturando tipos de váriaveis


```python
faturamento = 2000
custo = 500
lucro = faturamento - custo
print('O faturamento da loja foi: ' + ' R$ '+ str(faturamento) +' reais')
print('O custo da loja foi: ' +   ' R$ '  +   str(custo) + ' reais')
print('O lucro da loja foi: ' +   ' R$ '  +   str(lucro) + ' reais')
```

# Metodo Format 
  F-string

```python
faturamento = 3000
custo = 800
lucro = faturamento - custo

print('O faturamento da loja foi: {}. O custo da loja foi {}. Assim, o lucro da loja foi {}'.format(faturamento, custo, lucro))

# com f-strings dentro da chaves pode-se fazer operações
print(f'O faturamento da loja foi de {faturamento} e o lucro foi de {lucro} ')
```

# Mudanças de tipo de variável

```python
faturamento = int(input("Insira o faturamento: "))
custo = int(input("Insira o custo: "))
lucro = faturamento - custo
print(f"O Lucro da loja foi de R$ {lucro} reais")
```
