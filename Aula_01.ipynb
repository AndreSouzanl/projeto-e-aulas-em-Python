# Tipos de variaveis em Python
Uma variavel é um objeto, afinal tudo é um objeto em Python

int -> inteiro
string -> texto
float -> númerica com casas decimais (ponto flutuante)
bool ou boolean -> True ou False

obs: Váriaveis em Python não sao declaradas explicitamente, mas isso não significa que você deve ficar mudando o tempo todo.

obs: Cuidado com os nomes restritos ( Arquivo de Dowloads)
