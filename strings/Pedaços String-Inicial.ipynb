### Forma Básica

```python
precos = "Jan: 25, Fev: 27, Mar: 29"

precos_janeiro = precos[5:7]
print(f'O precos de Janeiro: {precos_janeiro}')
```

### Posição Inicial e Final

```python
preco_mar = precos[-2:]
print(f'Março: {preco_mar}')

preco_fev = precos[14:16]
print(f'Fevereiro : {preco_fev}')
preco_fev2 = precos[-11:-9]
print(f'Fevereiro 2: {preco_fev2}')
```

### Posição Inicial e Final com Step

```python
codigo = "1.2.3.4,5,1,2.3.4,7.9"

pedaco_cod = codigo[10:2:-2]
print(pedaco_cod)
```
